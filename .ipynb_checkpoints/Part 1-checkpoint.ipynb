```python
# Create test data

import numpy as np
from sklearn.model_selection import train_test_split

# Random Seed
# Random sampling is fine based on the number of instances in the data set 
np.random.seed(1998)
train_set, test_set = train_test_split(df, test_size=0.2, random_state=1998)
target = 'HadHeartAttack'

# stratified sampling on the target variable (HadHeartAttack)
# Given the target variable containts NaN values, I will delete the instances with NaN for HadHeartAttack from the dataset
print(df[target].value_counts())
print(df[target].isna().sum())

#Delete NaN rows from target variable
df = df.dropna(subset=[target])
print(df[target].isna().sum())
```

```python
# Create test data using train_test_split with stratify
np.random.seed(1998)
train_set, test_set = train_test_split(df, test_size=0.2, random_state=1998, stratify=df[target])
```

```python
# Preparing the data for ML

# Preparation 1 - Impute missing values for float64 data types 

from sklearn.experimental import enable_iterative_imputer  
from sklearn.impute import IterativeImputer

# median used for float cols because of the skew to 0 days in the dataset for physicalhealth, mentalhealth days etc.
# Increased max iteration from 10 to 30 due to cell crashing at max_iter=10 and max_iter=30
imputer = IterativeImputer(
    max_iter=30, 
    tol=1e-2, 
    random_state=42,
    min_value=0.0  
)
float_cols = ['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'HeightInMeters', 'WeightInKilograms', 'BMI']
train_set[float_cols] = imputer.fit_transform(train_set[float_cols])
test_set[float_cols] = imputer.transform(test_set[float_cols])
```

```python
# Preparation 2 - Convert Object data types into numerical data 
# 2.1 - Handle ordinal, nomial, binary, and pure data types

ordinal_features = ['GeneralHealth', 'AgeCategory', 'RemovedTeeth', 'HadDiabetes', 'SmokerStatus', 'ECigaretteUsage',
                   'TetanusLast10Tdap', 'CovidPos']

nominal_features = ['Sex', 'State', 'RaceEthnicityCategory', 'BMI', 'WeightInKilograms', 'HeightInMeters', 'SleepHours']

binary_features = [
    'PhysicalActivities', 'HadStroke', 'HadAsthma', 'AlcoholDrinkers', 
    'HadAngina', 'HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder', 
    'HadKidneyDisease', 'HadArthritis', 'DeafOrHardOfHearing', 
    'BlindOrVisionDifficulty', 'DifficultyConcentrating', 
    'DifficultyWalking', 'DifficultyDressingBathing', 'DifficultyErrands', 
    'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver', 'HighRiskLastYear'
]

#Features to remove:
remove = ['LastCheckupTime', 'ChestScan']
train_set = train_set.drop(columns=remove, errors='ignore')
test_set = test_set.drop(columns=remove, errors='ignore')
```

```python
# 2.2 - Giving structure to ordinal values

gen_health_rank = ['Poor', 'Fair', 'Good', 'Very good', 'Excellent']
age_rank = [
    'Age 18 to 24', 'Age 25 to 29', 'Age 30 to 34', 'Age 35 to 39', 
    'Age 40 to 44', 'Age 45 to 49', 'Age 50 to 54', 'Age 55 to 59', 
    'Age 60 to 64', 'Age 65 to 69', 'Age 70 to 74', 'Age 75 to 79', 
    'Age 80 or older'
]
teeth_rank = ['None of them', '1 to 5', '6 or more, but not all', 'All']
diabetes_rank = ['No', 'Yes, but only during pregnancy (female)', 'No, pre-diabetes or borderline diabetes', 'Yes']
smoker_rank = ['Never smoked', 'Former smoker', 'Current smoker - now smokes some days', 'Current smoker - now smokes every day']
ecig_rank = ['Never used e-cigarettes in my entire life', 'Not at all (right now)', 'Use them some days', 'Use them every day']
tet_rank = ['Yes, received Tdap', 'Yes, received tetanus shot, but not Tdap','Yes, received tetanus shot but not sure what type', 'No, did not receive any tetanus shot in the past 10 years']
covid_rank = ['No', 'Tested positive using home test without a health professional', 'Yes']
```

```python
# 2.3 - Transformation of all object data types 

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

#Impute and Encode 

# Ordinal object data pipeline
ord_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder(categories=[
        gen_health_rank, 
        age_rank, 
        teeth_rank,     
        diabetes_rank, 
        smoker_rank, 
        ecig_rank, 
        tet_rank, 
        covid_rank
    ])),
    ('scaler', StandardScaler())
])

# Nominal object data pipeline
nom_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='Unknown')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=True)),
])
```

```python
# 2.4 Preprocessing 

binary_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder()),
])

# 2. Update the Preprocessor
preprocessor = ColumnTransformer(transformers=[
    ('ordinal', ord_pipeline, ordinal_features),
    ('binary', binary_transformer, binary_features),
    ('nominal', nom_pipeline, nominal_features)
])


#Fit to the training set
X_train_final = preprocessor.fit_transform(train_set)
X_test_final = preprocessor.transform(test_set)
```

```python
# Feature Scaling 

# Ordinal features have already been scaled, check any skewness in the float data 

skewness = train_set.select_dtypes(include=['number']).skew().sort_values(ascending=False)
print("Skewness of numerical features:")
print(skewness)

```

```python
# Based on the skew of the data for PhysicalHealthDays, MentalHealthDays, BMI, and WeightinKilograms use a transformer

from sklearn.preprocessing import FunctionTransformer

skewed_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('log', FunctionTransformer(np.log1p, validate=True)), 
    ('scaler', StandardScaler())
])

normal_num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

highly_skewed = ['PhysicalHealthDays', 'MentalHealthDays', 'BMI', 'WeightInKilograms']
normal_num = ['HeightInMeters', 'SleepHours']

preprocessor = ColumnTransformer(transformers=[
    ('skewed', skewed_pipeline, highly_skewed),
    ('normal_num', normal_num_pipeline, normal_num),
    ('ordinal', ord_pipeline, ordinal_features),
    ('binary', binary_transformer, binary_features),
    ('nominal', nom_pipeline, nominal_features)
])

X_train_final = preprocessor.fit_transform(train_set)
X_test_final = preprocessor.transform(test_set)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# 1. Define the Pipeline with the correct model (LogisticRegression)
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000, class_weight='balanced')) # Add this!
])

# 2. Prepare the data - Map strings to numbers ONCE
y_train = train_set[target].map({'No': 0, 'Yes': 1})
y_test = test_set[target].map({'No': 0, 'Yes': 1})

X_train = train_set.drop(columns=[target])
X_test = test_set.drop(columns=[target])

# 3. Fit the model
# The pipeline handles the X transformation; you provide the numerical y
model_pipeline.fit(X_train, y_train)

# 4. Make predictions
y_pred = model_pipeline.predict(X_test)

# 5. Evaluate the results
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score

# Get the probabilities for the 'Yes' class
y_probs = model_pipeline.predict_proba(X_test)[:, 1]

# Calculate AUC
auc_score = roc_auc_score(y_test, y_probs)
print(f"ROC-AUC Score: {auc_score:.4f}")
```

```python
# Second Model - Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(
        n_estimators=100, 
        class_weight='balanced', 
        random_state=42,
        n_jobs=-1 
    ))
])

# 2. Train
rf_pipeline.fit(X_train, y_train)

# 3. Predict and Evaluate
y_pred_rf = rf_pipeline.predict(X_test)
y_probs_rf = rf_pipeline.predict_proba(X_test)[:, 1]

print("--- Random Forest Results ---")
print(classification_report(y_test, y_pred_rf))
print(f"ROC-AUC Score: {roc_auc_score(y_test, y_probs_rf):.4f}")
```
